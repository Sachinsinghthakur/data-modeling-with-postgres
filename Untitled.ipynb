writing and testing the function to get all json files path from given directory or folder.
*note* :- os.path.join gives all the folders path to our glob function with string(path) conconated with *json
##glob function will return all the files matching the pattern, in our case all the files ending with .json##

```python
def get_json_files_path(path):
    all_files_path = []
    for root,folder,files in os.walk(path):
        files=glob.glob(os.path.join(root,"*json"))
        for f in files:
            all_files_path.append(os.path.abspath(f))
    return all_files_path
```

```python
import numpy
from psycopg2.extensions import register_adapter, AsIs
def addapt_numpy_float64(numpy_float64):
    return AsIs(numpy_float64)
def addapt_numpy_int64(numpy_int64):
    return AsIs(numpy_int64)
def addapt_numpy_int32(numpy_int32):
    return AsIs(numpy_int32)

register_adapter(numpy.float64, addapt_numpy_float64)
register_adapter(numpy.int64, addapt_numpy_int64)
register_adapter(numpy.float64, addapt_numpy_int32)
```
